```python
# Import libraries
import csv
#import xlrd
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime as dt
```

```python
# Read world population data
fileData1 = 'ex2-2/world-population.xlsm'
population = pd.read_excel(fileData1)


print(population.head())
```

### World Population data
#### Line and Step graph

```python
X = population['Year']
Y = population['Population']
plt.title('World Population Growth - Line')
plt.plot(X, Y)
plt.show()
```

```python
plt.title('World Population Growth - Stepped')
plt.step(X, Y)
plt.show()
```
